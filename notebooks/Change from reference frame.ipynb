```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the images
image1 = cv2.imread('image1.jpg')
image2 = cv2.imread('image2.jpg')

# Convert the images to grayscale
image1_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
image2_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Calculate the absolute difference between the images
difference = cv2.absdiff(image1_gray, image2_gray)

# Threshold the difference image
_, difference = cv2.threshold(difference, 25, 255, cv2.THRESH_BINARY)
```

```python
# Dilate the difference image to amplify the changes
kernel = np.ones((1,1), np.uint8)
difference = cv2.dilate(difference, kernel, iterations=3)
```

```python
# Check if any pixels are non-zero (movement detected)
if cv2.countNonZero(difference) > 0:
    print("Motion detected")
else:
    print("No motion detected")
```

```python
# Display the difference image using Matplotlib
plt.imshow(difference, cmap='gray')
plt.show()
```

```python
# Display the difference image
cv2.imshow('Difference', difference)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
